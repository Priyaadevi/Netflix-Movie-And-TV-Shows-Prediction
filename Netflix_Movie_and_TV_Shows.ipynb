```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df["title"], df["release_year"], marker='o', linestyle='-', linewidth=2)
# Customize the plot
plt.title("Release year vs title")
plt.xlabel("title")
plt.ylabel("release_year")
# Show the plot
plt.show()
```

```python
# Plot the data
plt.figure(figsize=(6, 6))
plt.bar(df["title"], df["release_year"], color='skyblue')
# Customize the plot
plt.title("Release year vs title")
plt.xlabel("title")
plt.ylabel("release_year")
# Show the plot
plt.show()
```
